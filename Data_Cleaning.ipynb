```python
import pandas as pd
import scipy
```

```python
df = pd.read_csv("planets.csv")
```

```python
df.head()
```

```python
cleaned_planets = df.dropna()
```

```python
cleaned_planets['Semi-Major Axis (km)'] = cleaned_planets.loc[:, 'Semi-Major Axis'] * 1.496E8
```

```python
cleaned_planets
```

```python
cleaned_planets['Semi-Major Axis(m)'] = cleaned_planets["Semi-Major Axis (km)"] * (10**3)
```

```python
cleaned_planets
```

```python
cleaned_planets["Stellar Mass (kg)"] = cleaned_planets["Stellar Mass (solar Mass)"] * 1.989E30
```

```python
cleaned_planets
```

```python
cleaned_planets["Orbital Velocity(m/s)"] = ((6.67E-11* cleaned_planets["Stellar Mass (kg)"])/cleaned_planets["Semi-Major Axis(m)"])**0.5
```

```python
cleaned_planets
```

```python
cleaned_planets.drop(columns=["Semi-Major Axis (km)"])
```

```python
cleaned_planets['Orbital Velocity (proportional)'] = cleaned_planets['Orbital Velocity(m/s)'] / (10**4)
```

```python
cleaned_planets
```

```python
cleaned_planets["Semi-Major Axis (proportional)"] = cleaned_planets['Semi-Major Axis (km)']/(10**5)
```

```python
cleaned_planets
```

```python
cleaned_planets.to_csv("cleaned_planets.csv")
```
