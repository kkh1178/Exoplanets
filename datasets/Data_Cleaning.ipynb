```python
import pandas as pd
import scipy
```

```python
df = pd.read_csv("planets.csv")
```

```python
df.head()
```

```python
cleaned_planets = df.dropna()
```

```python
cleaned_planets['Semi_Major_Axis_km'] = cleaned_planets.loc[:, 'Semi_Major_Axis_AU'] * 1.496E8
```

```python
cleaned_planets
```

```python
cleaned_planets['Semi_Major_Axis_m'] = cleaned_planets["Semi_Major_Axis_km"] * (10**3)
```

```python
cleaned_planets
```

```python
cleaned_planets["Stellar_Mass_kg"] = cleaned_planets["Stellar_Mass_solar_Mass"] * 1.989E30
```

```python
cleaned_planets
```

```python
cleaned_planets["Orbital_Velocity_ms"] = ((6.67E-11* cleaned_planets["Stellar_Mass_kg"])/cleaned_planets["Semi_Major_Axis_m"])**0.5
```

```python
cleaned_planets
```

```python
cleaned_planets.drop(columns=["Semi_Major_Axis_km"])
```

```python
cleaned_planets['Orbital_Velocity_proportional'] = cleaned_planets['Orbital_Velocity_ms'] / (10**4)
```

```python
cleaned_planets
```

```python
cleaned_planets["Semi_Major_Axis_proportional"] = cleaned_planets['Semi_Major_Axis_km']/(10**5)
```

```python
cleaned_planets
```

```python
cleaned_planets['Planet_Radius_km'] = cleaned_planets['Planet_Radius_jupiter_Radii'] * 71492
cleaned_planets['Planet_Radius_Earth_Radii'] = cleaned_planets['Planet_Radius_km'] / 6371
cleaned_planets
```

```python
cleaned_planets["Distance_to_Star_ly"] = cleaned_planets["Distance_to_Star_pc"] * 3.26
cleaned_planets["minimum_Travel_Time_yr"] = ((cleaned_planets['Distance_to_Star_ly'] * 9.461e12) / 73.61)/(1314000)
cleaned_planets
```

```python
cleaned_planets.to_csv("cleaned_planets.csv")
```
